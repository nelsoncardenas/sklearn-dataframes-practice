## Transformations plan

1. NaN correction: replace "?" for NaN.
2. Imputer: impute missing values mode and median.
3. Categorical encoder: OHE for categorical.
4. Scaler: standarization for numerical columns.
5. NaN handler: drop columns with more than 40% of NaNs.

```python
df.replace("?", np.nan, inplace = True)
```

```python
missing_vals = df.isna().sum().to_frame(name="num_nans")
nrows = df.shape[0]
missing_vals = missing_vals.assign(frac_nans = missing_vals["num_nans"]/nrows)
missing_vals
```

```python
missing_vals[missing_vals["frac_nans"] >= 0.4].index.values.tolist()
```

```python
num_uniques = df.nunique().to_frame(name="num_uniques")
missing_vals = num_uniques.assign(
    frac_uniques=num_uniques["num_uniques"] / nrows
)
missing_vals = missing_vals[
    ~missing_vals.index.isin(["incident_date", "policy_bind_date"])
]
columns_to_drop = missing_vals[
    missing_vals["frac_uniques"] >= 0.9
].index.values.tolist()
columns_to_drop
```
